```python
import pandas as pd
from sklearn import linear_model
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
```

```python
df.head()
```

```python
df.isna()
```

```python
df['per_capita_income'] = df['per capita income (US$)']
```

```python
df.head()
```

```python
df = df.drop('per capita income (US$)',axis=1)
```

```python
df.head()
```

```python
plt.xlabel('year')
plt.ylabel('per_capita_income')
plt.scatter(df.year,df.per_capita_income)
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df['per_capita_income'])
```

```python
ans = reg.predict([[2020]])
```
